```python
# time模块提供了与时间相关的操作，使用import加载模块
import time 

n = 5
while n > 0:
    print (n)
    time.sleep(1) # 让程序停止一段时间,可以设置int或float，单位为秒
    n -= 1 # 在循环中必须修改条件变量，在这里条件变量是n

print ('Blastoff')
print (n)
```

```python
while True:
    line = input('> ')
    if line == 'quit':
        break #直接停止并跳出循环，结束整个循环
    if line == 'sleep':
        continue # 结束当前循环，不执行后续语句，直接进入下一轮循环
    print (line)

print ('out of the while')
```

```python
# pass语句，空语句不执行任何事，用来保持程序结构的完整性
for letter in 'Ishahduishiu':
    if letter == 'h':
        pass
        print ("这是pass块")
    print ('当前字母: ', letter)

print ('完成!')
```

```python
# 打印九九乘法表

n = 1

while n < 10:
    i = 1
    while i <= n:
        print (i, '*', n, '=', i*n, end = '    ')
        i += 1
    print()
    n += 1
```

```python
# 寻找一个数的所有因数
n = int(input())
for i in range (1, int(n ** 0.5)):
    if ( n % i == 0):
        print (i, ' ', n / i)
```
