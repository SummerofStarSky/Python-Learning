```python
# List是用一个变量名定义一个集合，且需要将集合中的元素都列举出来
set1 = [1, 2, 3, 4, 5]
set2 = ['red', 'blue', 'green', 'white', 'black']
set3 = ['purple', 9, 3.141]
set4 = [[1, 2, 3, 4, 5], ['red', 'blue', 'green', 'white', 'black'], ['purple', 9, 3.141]]
set5 = []
print (set1)
print (set2)
print (set3)
print (set4)
print (set5)
```

```python
# list可以和for联合使用,非常便利
set1 = [1, 2, 3, 4, 5]
for e in set1:
    print (e)
```

```python
# 使用索引操作列表中的元素，可以获取或修改，索引也是下标，由左到右从0开始计
set2 = ['red', 'blue', 'green', 'white', 'black']
# 反向索引 -5 -4 -3 -2 -1
# 正向索引 0 1 2 3 4
print (set2[0])
```

```python
# 使用len()来获取list的长度
set2 = ['red', 'blue', 'green', 'white', 'black']
print (len(set2))

# 搭配使用len()和range()
for i in range(len(set2)):
    colour = set2[i]
    print ('I like',colour)
```

```python
# 修改list中的元素
set1 = [1, 2, 3, 4, 5]
for i in range(len(set1)):
    set1[i] *= 2
print (set1)
```

```python
# list常见方法
# len(list) 求长度
# max(list) 返回列表中的最大值
# min(list) 返回列表中的最小值
# list(seq) 将元组转换为列表
set1 = [1, 2, 3, 4, 5]
print (len(set1))
print (min(set1))
print (max(set1))

# 两个list之间的操作：列表 + 列表 代表连接；列表 * 整数 代表重复
set3 = ['purple', 9, 3.141]
a = set1 + set3
b = set1 * 4
print (a)
print (b)
```

```python
# list添加、删除和排序元素
set5 = []
print (set5)

set5.append('hello') # 使用append()添加元素
set5.append('99')
set5.append('bye')
print (set5)

set5.remove('99') # 使用remove()删除元素
print (set5)

set5.sort() # 使用sort()对列表中的元素进行排序
print (set5)
```

```python
# 使用input()向列表中输入元素
set6 = list() # 等价于 set6 = [], 即初始化列表
while True :
    num = input('Enter a number: ') # 输入数字时，应输入一个后立即回车再输入下一个，与C++中的输入方式是完全不同的
    if num == 'done':
        break
    value = float(num) 
    set6.append(value)
print (set6) 
average = sum(set6) / len(set6) # sum()可以对数字列表求和
print ('Average', average)
```
