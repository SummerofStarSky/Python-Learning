```python
#模块Module
'''
一个Python文件，
里面有类，变量，函数
[from 模块名] import [模块 | 类 | 函数] [as 别名]
'''

import time as t
# 导入Python内置模块
# 使用.来调用任一功能
print("hello")
t.sleep(5) #程序睡眠5秒
print("hello")

from time import sleep
# 单独导入一个具体的功能，而不是整个块
print("hello")
sleep(5)
print("hello")

from time import * 
# *表示全部的功能导入
print("hello")
sleep(5)
print("hello")
```

```python
# 自定义模块
# 新建一个Python文件，定义函数，使用时直接import + 文件名
# 模块调用时会自动执行该模块所有代码
def test(a, b):
    print(a + b)

# 表示只有程序是被直接执行而不是被导入时才会执行
if __name__ == "__main__":
    test(1,2)

# 如果一个模块文件中有__all__变量，当使用from XXX import * 导入时， 只能导入这个列表中的元素
def test_a(a, b):
    print(a + b)

def test_b(a, b):
    print(a - b)

__all__ = ["test_a"]
```
