```python
# exception异常
#捕获异常

#基本捕获语法
try:
    f = open("abc.txt", "r", encoding="UTF-8")
except:
    print("Error")
    f = open("abc.txt", "w", encoding="UTF-8")
    
#捕获指定的异常
try:
    print(name)
except NameError as e:
    print("出现了变量未定义的异常")
    print(e)

#捕获多个异常
try:
    print(name)
except (NameError, ZeroDivisionError) as e:
    print("出现了变量未定义 或除以0的异常")

#捕获所有异常
#第一种基本语法是一种
#另外一种如下
try:
    f = open("D:/123.txt", "r", encoding="UTF-8")
except Exception as e:
    print("Error")
else: #可选择写不写，如果没有异常就执行
    print("No Error!")
finally: #无论有没有异常都要执行，也可以选择要不要写
    f.close()
```
