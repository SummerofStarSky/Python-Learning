```python
#传参方式
#数据的传递， 数据不确定， 计算逻辑确定
#位置参数
def user_info(name, age, gender):
    print(f"姓名是{name}, 年龄是:{age}, 性别是:{gender}")

user_info("小明", 20, "男")
#关键字传参
user_info(name="小王", age=11, gender="女")
user_info(age=11, gender="女", name="小王")
user_info("小王", age=11, gender="女")   
          
#缺省传参
def user_info(name, age, gender="男"):
    print(f"姓名是:{name}, 年龄是:{age}, 性别是:{gender}")
#可设置默认值，统一放在最后
user_info('Valor',100)

#不定长参数

#不定长 - 位置不定长， *号
#不定长定义的形式参数会作为元组存在， 接收不定长数量的参数传入
def user_info(*args):
    print(f"args参数的类型是: {type(args)}, 内容是:{args}")
user_info(1, 2, 3, '小明', '男孩')

#不定长 - 关键字不定长， **号
#不定长定义的形式参数会作为字典存在， 接收不定长数量的参数传入
def user_info(**kwargs):
    print(f"kwargs参数的类型是: {type(kwargs)}, 内容是:{kwargs}")

user_info(name='小王', age=11, gender='男孩')

          
```

```python
#函数参数传递也可以传函数
#计算逻辑的传递， 数据确定，计算逻辑不确定
def test_func(compute):
    result = compute(1, 2)
    print(result)
    print(type(compute))

def compute(x, y):
    return x + y

test_func(compute)
```

```python
#lambda, 匿名函数
'''
def关键字，可以定义有名称的函数，可重复利用
lambda关键字，可以定义无名字的函数，只可临时使用一次
'''
def test_func(compute):
    result = compute(1, 2)
    print(result)
    print(type(compute))

# lambda 传入参数： 函数体（一行代码）
# lambda 自动return结果

test_func(lambda x,y: x + y)
```
