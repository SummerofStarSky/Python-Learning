```python
#文件操作

#打开
f = open("D:/测试.txt", "r", encoding = "UTF-8")
print(type(f))

#读

#文件对象.read(num), 可以选择读取多少字节的内容,不写则全读
print(f.read(10))
print(f.read()) #连续多个read时，会接着上一个进行读取

#文件对象.readlines()
lines = f.readlines()
print(type(lines))
print(lines)
#不管哪种方法，如果连续，都会接着上一个读取的末尾继续读取

#文件对象.readline()
line1 = f.readline()
line2 = f.readline()
print(line1)
print(line2)

#for循环来读取文件
for line in f:
    print(line)

#关闭
f.close()

#with open("路径、文件名"，"mode", encoding = "UTF-8") as f， 可以自动关闭文件
with open("测试.txt", "r", encoding = "UTF-8") as f:
    for line in f:
        print(line)

#黑马程序员的案例练习
#方法一，直接用method计数
with open("word.txt", "r", encoding="UTF-8") as f:
    content = f.read()
    count1 = content.count("itheima")

#方法二，一行行读
count2 = 0
with open("word.txt", "r", encoding="UTF-8") as f:
    for line in f:
        line = line.strip() #处理开头和结尾的换行符和空格
        words = line.split(" ")
        for word in words:
            if word == "itheima":
                count2 += 1 
    
```

```python
#文件写入
#f.write(), 内容写入
with open("test.txt", "w", encoding="UTF-8") as f:
    f.write("Hello World!") # 不存在的文件会自动创建新文件，文件存在会将文件原内容全部清空
    # 只是将内容写入运行内存中，而不是硬盘中
    f.flush()
    # 将内容写入到硬盘中
    # 不用flush，close（）自带flush（）功能
    
#f.flush(), 将文件写入硬盘

#文件的追加操作，append（）
# 文件不存在，创建新文件；文件存在，接在原内容末尾继续写入
with open("text.txt", "a", encoding="UTF-8") as f:
    f.write("I am a beginner.")
    f.write("\nI want to contribute to my country")
```
