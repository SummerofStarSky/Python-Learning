```python
#set不可重复，可修改，无顺序->不可索引
s1 = {'S', 'S', 'E', 'D'}
s2 = set()
s3 = {'e', 'S', 'E', 'd'}

print (s1)
print (s2)

#add() :加入新元素
s1.add('R')
print(s1)

#remove() :删除特定元素
s1.remove('R')
print(s1)

#求差集，顺序有影响
print(s1.difference(s3)) 
print(s3.difference(s1))

#pop() : 随机删除一个元素
print(s1.pop())

#union() :求并集
print(s1.union(s3))
print(s3.union(s1))

#intersection : 求交集
print(s1.intersection(s3))
print(s3.intersection(s1))

#clear() :清空集合
print(s1.clear())
```

```python
# dictionary
'''
key: value 键值对
与map，hash同意
key 不可变类型：list，set，dict
value 可变类型
不支持下表索引
'''

dict1 = {"赖":636, "李":636, "卢":629, "程":631, "钟":629, "蔡":671, "刘":503, "段":617}
print(dict1)
print(type(dict1))

# key不能重复，否则后一个值会覆盖前一个值

#访问
print(dict1["赖"])

#新增
dict1["曾"] = 641
print(dict1)

#更新
dict1["曾"] = 654
print(dict1)

#删除
score = dict1.pop("曾")
print(dict1)
print(score)

#清空, clear
dict1.clear()
print(dict1)


dict1 = {"赖":636, "李":636, "卢":629, "程":631, "钟":629, "蔡":671, "刘":503, "段":617}

#获取全部的key
keys = dict1.keys()
print(keys)

#遍历字典
#方式1
for key in keys:
    print(key)
    print(dict1[key])
#方式2
for key in dict1:
    print(key)
    print(dict1[key])
    
#统计字典内的元素数量， len（）函数
num = len(dict1)
print(num)

         
```
