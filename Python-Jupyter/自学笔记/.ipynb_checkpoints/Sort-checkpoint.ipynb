```python
def bubble_sort (arr):
    n = len(arr)
    for i in range (n - 1):
        for j in range (n - 1 - i):
            if arr[j] > arr[j+1]:
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp

num = int(input("Enter the number of numbers that need to be sorted:"))
arr = []

for k in range (num):
    element = int(input("Enter the number"))
    arr.append(element)

bubble_sort (arr)

print (arr)
```

```python
def insert_sort (arr):
    n = len(arr)
    for i in range (1 , n):
        temp = arr[i]
        for j in range (i-1 , -1, -1):
            if temp < arr[j]:  #向左比较，如果左边数更大把它向右移动
                arr[j+1] = arr[j]
                if j == 0:    #如果移完后，发现j=0，就没有必要继续比较了，直接把temp赋给arr[0],并结束这层循环
                    arr[j] = temp
                    break
            else: 
                arr[j+1] = temp #如果发现arr[j] <= temp, 则直接将temp移到arr[j+1]的位置
                break
                
num = int(input("Enter the number of numbers that need to be sorted:"))
arr = []

for k in range (num):
    element = int(input("Enter the number"))
    arr.append(element) # alevel计算机paper4考试不允许一直用append()定义数组

bubble_sort (arr)

print (arr) 
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range (0,n):
        index = i
        for j in range (i+1,n):
            if arr[j] < arr[index]:
                index = j
        if index != i:
            temp = arr[i]
            arr[i] = arr[index]
            arr[index] = temp

num = int(input("Enter the number of numbers that need to be sorted:"))
arr = []

for k in range (num):
    element = int(input("Enter the number"))
    arr.append(element)

selection_sort (arr)

print (arr) 
```

```python
def selection_sort(arr): # 同时寻找并交换min和max
    n = len(arr)
    left = 0
    right = n-1
    while left < right:
        min_index = left
        max_index = right
        for i in range (left,right+1):
            if arr[i] < arr[min_index] :
                min_index = i
            if arr[i] > arr[max_index] :
                max_index = i
        if min_index != left:
            temp1 = arr[min_index]
            arr[min_index] = arr[left]
            arr[left] = temp1
        if max_index == left:
            max_index = min_index # 当找到的maxima真好在left时，要找到正确的index
        if max_index != right:
            temp2 = arr[right]
            arr[right] = arr[max_index]
            arr[max_index] = temp2
        left = left + 1
        right = right - 1

num = int(input("Enter the number of numbers that need to be sorted:"))
arr = []

for k in range (num):
    element = int(input("Enter the number"))
    arr.append(element)

selection_sort (arr)

print (arr) 
```
