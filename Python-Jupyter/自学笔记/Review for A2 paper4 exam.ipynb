```python
# linear search
def linear_search(items, target):
    for index in range(len(items)):
        if items[index] == target:
            return index
    return -1

names = ["Ali", "Beth", "Chen"]
position = linear_search(names, "Chen")
print (position)
```

```python
# Binary search
def binary_search(items, target):
    low = 0
    high = len(items) - 1

    while low <= high:
        mid = (low + high) // 2
        if items[mid] == target:
            return mid
        elif items[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

numbers = [3, 8, 12, 19, 25, 31]
print (binary_search(numbers, 19))
```

```python
# Bubble sort by marks descending
def bubble_sort_by_mark(records):
    n = len(records)
    for i in range(n - 1):
        swapped = False
        for j in range(n - 1 - i):
            if records[j][2] < records[j + 1][2]:
                records[j], records[j + 1] = records[j + 1], records[j]
                swapped = True
        if not swapped:
            break

students = [
    ["S001", "Ali", 72],
    ["S002", "Beth", 85],
    ["S003", "Chen", 64]
]
bubble_sort_by_mark(students)
print (students)
```

```python
# insertion sort ascending
def insertion_sort(numbers):
    for i in range(1, len(numbers)):
        current = numbers[i]
        j = i - 1
        while j >= 0 and numbers[j] > current:
            numbers[j + 1] = numbers[j]
            j = j - 1
        numbers[j + 1] = current

values = [7, 3, 9, 1]
insertion_sort(values)
print(values)
```

```python
# Recursion
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))

''' checklist
base case
每次调用是否更接近base case
return value是否正确传回
大输入会不会造成stack overflow
'''
```

```python
# Stack
class Stack:
    def __init__(self, max_size):
        self.items = []
        self.max_size = max_size

    def is_empty(self):
        return len(self.items) == 0

    def is_full(self):
        return len(self.items) == self.max_size

    def push(self, item):
        if self.is_full():
            return False
        self.items.append(item)
        return True

    def pop(self):
        if self.is_empty():
            return None
        return self.items.pop()

    def peek(self):
        if self.is_empty():
            return None
        return self.items[-1]
        
```

```python
# Queue
class Queue:
    def __init__(self, max_size):
        self.items = []
        self.max_size = max_size

    def is_empty(self):
        return len(self.items) == 0

    def is_full(self):
        return len(self.items) == self.max_size

    def enqueue(self, item):
        if self.is_full():
            return False
        self.items.append(item)
        return True

    def dequeue(self):
        if self.items.is_empty():
            return None
        return self.items.pop(0)
```

```python
# Circular Queue
class CircularQueue:
    def __init__(self, size):
        self.data = [None] * size
        self.front = 0
        self.rear = -1
        self.count = 0
        self.size = size

    def is_empty(self):
        return self.count == 0

    def is_full(self):
        return self.count == self.size

    def enqueue(self, item):
        if self.is_full():
            return False
        self.rear = (self.reat + 1) % self.size
        self.data[self.rear] = item
        self.count += 1
        return True

    def dequeue(self):
        if self.is_empty():
            return None
        item = self.data[self.front]
        self.data[self.front] = None
        self.front = (self.front + 1) % self.size
        self.count -= 1
        return item
        
```

```python
# class and constructors
class Student:
    def __init__(self, student_id, name, mark):
        self.__student_id = student_id
        self.__name = name
        self.__mark = mark

    def get_student_id(self):
        return self.__student_id

    def get_name(self):
        return self.__name

    def get_mark(self):
        return self.__mark

    def set_mark(self, new_mark):
        if 0 <= new_mark <= 100:
            self.__mark = new_mark
            return True
        return False

    def calculate_grade(self):
        if self.__mark >= 80:
            return 'A'
        elif self.__mark >= 70:
            return 'B'
        elif self.__mark >= 60:
            return 'C'
        else:
            return 'U'

student = Student("S001", "Ali", 72)
print(student.get_name(), student.calculate_grade())
```

```python
# Inheritance
class Account:
    def __init__(self, account_number, balance):
        self.__account_number = account_number
        self.__balance = balance

    def get_balnace(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

class SavingAccount(Account):
    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.__interest_rate = interest_rate

    def calculate_interest(self):
        return self.get_balance() * self.__interest_rate
```
