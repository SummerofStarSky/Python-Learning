```python
# List是用一个变量名定义一个集合，且需要将集合中的元素都列举出来
set1 = [1, 2, 3, 4, 5]
set2 = ['red', 'blue', 'green', 'white', 'black']
set3 = ['purple', 9, 3.141]
set4 = [[1, 2, 3, 4, 5], ['red', 'blue', 'green', 'white', 'black'], ['purple', 9, 3.141]]
set5 = []
print (set1)
print (set2)
print (set3)
print (set4)
print (set5)
```

```python
# list可以和for联合使用,非常便利
set1 = [1, 2, 3, 4, 5]
for e in set1:
    print (e)
```

```python
# 使用索引操作列表中的元素，可以获取或修改，索引也是下标，由左到右从0开始计
set2 = ['red', 'blue', 'green', 'white', 'black']
# 反向索引 -5 -4 -3 -2 -1
# 正向索引 0 1 2 3 4
print (set2[0])
```

```python
# 使用len()来获取list的长度
set2 = ['red', 'blue', 'green', 'white', 'black']
print (len(set2))

# 搭配使用len()和range()
for i in range(len(set2)):
    colour = set2[i]
    print ('I like',colour)
```

```python
# 修改list中的元素
set1 = [1, 2, 3, 4, 5]
for i in range(len(set1)):
    set1[i] *= 2
print (set1)
```

```python
# list常见方法
# len(list) 求长度
# max(list) 返回列表中的最大值
# min(list) 返回列表中的最小值
# list(seq) 将元组转换为列表
set1 = [1, 2, 3, 4, 5]
print (len(set1))
print (min(set1))
print (max(set1))

# 两个list之间的操作：列表 + 列表 代表连接；列表 * 整数 代表重复
set3 = ['purple', 9, 3.141]
a = set1 + set3
b = set1 * 4
print (a)
print (b)
```

```python
# list添加、删除和排序元素
set5 = []
print (set5)

set5.append('hello') # 使用append()添加元素
set5.append('99')
set5.append('bye')
print (set5)

set5.remove('99') # 使用remove()删除元素
print (set5)

set5.sort() # 使用sort()对列表中的元素进行排序
print (set5)
```

```python
# 使用input()向列表中输入元素
set6 = list() # 等价于 set6 = [], 即初始化列表
while True :
    num = input('Enter a number: ') # 输入数字时，应输入一个后立即回车再输入下一个，与C++中的输入方式是完全不同的
    if num == 'done':
        break
    value = float(num) 
    set6.append(value)
print (set6) 
average = sum(set6) / len(set6) # sum()可以对数字列表求和
print ('Average', average)
```

```python
# 其他列表常用函数

# list.count()统计某个元素在列表中出现的元素
set7 = [1, 1, 4, 67, 3, 1, 5, 45, 4, 1]
num1 = set7.count(1)
print (num1)

# list.extend(seq) 在列表末尾一次性追加另一个序列中的多个值（用新列表扩展原列表）
set1 = [1, 2, 3, 4, 5]
set8 = set7.extend(set1)
print (set8)
# 有点问题，应该是把元组加入列表？

# list.index()从列表中找出某个值第一个匹配项的索引位置, 注意是第一个匹配项，列表与真正的集合
place1 = set7.index(1)
place4 = set7.index(4)
print (place1)
print (place4)

# list.insert(index, object) 将对象插入列表对应索引位置，原位置及以后的元素向后位移一位
set7.insert(3, 0)
print (set7)

# list.pop([index = -1]) 移除列表中的一个元素，默认是最后一个元素，并且返回该元素的值
lastvalue = set7.pop()
print (lastvalue)
print (set7)
value0 = set7.pop(0) # 括号内填入index可以删除对应元素并返回其值
print (value0)
print (set7)

# list.remove() 删除列表中某一个值的第一个匹配项
set7.remove(1)
print (set7)

# list.reverse() 反向列表中的元素
set7.reverse()
print (set7)
# list.[::-1]也可以用来反向列表
print (set7[::-1])
# 这里的输出将之前反向过的列表重新反向，得到原列表

# list.sort(cmp=None, key=None, reverse=False) 对原列表进行排序 cmp=None已在Python3中淘汰；
# key=None表示使用默认的比较方式，即比较元素本身的大小，但是允许key等于其他自定义或内置的函数作为比较规则； 
# reverse参数用来确定结果是升序或降序，默认值是False代表升序，True代表降序
set9 = set7
set7.sort(key=None, reverse=False)
print (set7)
set9.sort(key=None, reverse=True)
print (set9)
```

```python
# list的分片（Slice）
cities = ["San Diego", "Prague", "Cape Town", "Tokyo", "Melbourne"]

a1 = cities[1:3] # 使用冒号进行分片，两个索引（包括前一个，不包括后一个）
print (a1)

a2 = cities[:4] # 只使用一个索引
print (a2)

a4 = cities[4:]
print (a4)

a3 = cities[-3:] # 使用负索引
print (a3)

a5 = cities[:-3]
print (a5)

a6 = cities[1] # 取一个元素
print (a6)

a7 = cities[1:4:1] # list.[start, stop, step] 'start'是起始位置，'stop'是终止位置，'step'是步长指的是取元素时的空格，从左往右取，默认值为1
print (a7)
a8 = cities[1:4:2]
print (a8)
a9 = cities[-1:-5:-2] #当step为负值是则为从右往左取，start的值也应大于stop的值，使得整个过程从右往左进行
print (a9)
a10 = cities[3:0:-1]
print (a10)
```
