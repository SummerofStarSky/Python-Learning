```python
# 用class来实现stack
class Stack():
    
    def __init__(self):   # initialize初始化，创建变量是会自动执行，第一个执行
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

    def peak(self):
        if not self.is_empty():
            return self.items[0]

    def size(self):
        return len(self.items)

s = Stack()
s.push(5)
s.push(9)
print(s.pop())  # 9
print(s.peak()) # 5
```

```python
# circular queue
class CircularQueue():

    def __init__(self,size):
        self.size = size
        self.queue = [None] * size
        self.front = -1
        self.rear = -1

    def enqueue(self,item):
        if (self.rear + 1) % self.size == self.front:
            print ("Queue is full!")
            return
        if self.front == -1:
            self.front = 0
        self.rear = (self.rear + 1) % self.size
        self.queue[self.rear] = item

    def dequeue(self):
        if self.front == -1:
            print ("Queue is empty!")
            return None
        temp = self.queue[self.front]
        if self.front == self.rear :
            self.front = -1
            self.rear = -1
        else: 
            self.front = (self.front + 1) % self.size
        return temp

    def get_front(self):
        if self.front == -1:
            print ("Empty!")
            return
        return self.queue[self.front]
        
q = CircularQueue(6)
q.enqueue(98)
q.enqueue(65)
print (q.dequeue())
```

```python
# linked list
class Node():
    def __init__(self,data = None):
        self.data = data
        self.next = None

head = None

def Add_node(head, item):
    
    new_node = Node(item)
    
    if head == None:
        return new_node
        
    current_node = head
    
    while current_node.next != None:
        current_node = current_node.next
    
    current_node.next = new_node
    return head

for i in range (4):
    head = Add_node(head, i * 2)

cur = head
while cur != None:
    print(cur.data)
    cur = cur.next

def delete_node(head, item):
    if head == None:
        return head

    if head.data == item:
        return head.next

    current_node = head
    previous_node = None
    while current_node.next != None and current_node.data < item:
        previous_node = current_node
        current_node = current_node.next

    if current_node.next != None and current_node.data == item:
        previous_node.next = current_node.next
        del current_node
    else:
        print (f"The target value {item} is not found!")
    return head

head = delete_node(head, 5)
head = delete_node(head, 4)

cur = head
while cur != None:
    print(cur.data)
    cur = cur.next

```

```python
class Stack:

    def __init__(self,size=10):
        self.capacity = size
        self.items = [None] * size
        self.TopPointer = -1
        
    def push(self, item):
        if self.TopPointer >= self.capacity:
            return False
        self.TopPointer += 1
        self.items[TopPointer] = item
        return True

    def pop(self):
        if self.is_empty():
            print("It is empty!")
            return None
        item = self.items[TopPointer]
        TopPointer -= 1
        return item

    def is_empty(self):
        if self.TopPointer == -1:
            return True
        return False

       # return self.TopPointer == -1
            
    def is_full(self):
        return self.TopPointer == self.capacity - 1

    def peek(self): #return top element of the stack
        if self.is_empty():
            print("It is empty!")
            return None
        return self.items[TopPointer]

    def display(self):
        if self.is_empty():
            print("It is empty!")
            return None
        for i in range(TopPointer, -1, -1):
            # print (self.items[i], end = ' ')
            print (f"{self.items[i]} ")
            
```
