```python
from datetime import date

# 创建日期：年, 月, 日
today = date.today()        # 今天
my_birthday = date(2008, 6, 11)

print(today)
print(my_birthday)

# 可以单独取出年、月、日
print(today.year)
print(today.month)
print(today.day)

# 计算两个日期相差天数
delta = today - my_birthday
print(delta.days)

# 比较日期
print(today > my_birthday)   # True / False

# 日期 → 字符串
s = today.strftime("%Y-%m-%d")
print (s)

# 字符串 → 日期
d = date.strptime("2025-12-25", "%Y-%m-%d")
print (d)
print (type(d))
```
