```python
class Node(): #blue map
    def __init__(self, data = None):
        self.data = data
        self.left = None
        self.right = None

    def display_ordered(self):
        if self.left != None:
            self.left.display_ordered()
        print (self.data)
        if self.right != None:
            self.right.display_ordered()

    def display_pre(self):
        print (self.data)
        if self.left != None:
            self.left.display_pre()
        if self.right != None:
            self.right.display_pre()

    def display_post(self):
        if self.left != None:
            self.left.display_post()
        if self.right != None:
            self.right.display_post()
        print (self.data)

    def min_node(self):
        min_value = self.data
        while self.left is not None:
            self = self.left
            min_value = self.data
        return min_value

    def delete_min_node(self):
        current = self
        previous = None
        while current.left is not None:
            previous = current
            current = current.left

        if previous == None:
            self.right = current.right
        else:
            previous.left = current.right
            
def add_node(root, item):
    if root == None:
        return Node(item)

    current = root
    while True:
        if item < current.data:
            if current.left is not None:
                current = current.left
            else:
                current.left = Node(item)
                break
        else: 
            if current.right is not None:
                current = current.right
            else:
                current.right = Node(item)
                break

    return root

bst = Node('G') #instanisation, object = instance, 保存地址去访问
# 引用类型，非指针类型，指针需要解引用
values = ['d', 's', 'r', 'y', 'x']
for value in values:
    bst = add_node(bst, value)

def find_node(root, value):
    current = root

    while current is not None:
        if current.data == value:
            return current
        elif current.data > value:
            current = current.left
        else:
            current = current.right

    return None

print (find_node(bst, 'G'))
print (find_node(bst, 'd'))
print (find_node(bst, 's'))
print (find_node(bst, 'r'))
print (find_node(bst, 'y'))
print (find_node(bst, 'x'))

bst.display_ordered()
bst.display_pre()
bst.display_post()

print (bst.min_node())

def delete_node(root, value):
    current_node = root
    previous_node = None
    
    if root == None:
        print ("There is no value in the tree!")
        return root

    while current_node is not None:
        if current_node.data == value:
            if current_node.left is None and current_node.right is None:
                if previous_node is None:
                    return None
                if previous_node.left == current_node:
                    previous_node.left = None
                else:
                    previous_node.right = None
                return root
            elif current_node.left is not None and current_node.right is not None:
                current_node.data = current_node.right.min_node()
                current_node.right.delete_min_node()
                return root
            elif current_node.left is not None and current_node.right is None:
                if previous_node is None:
                    return current_node.left
                if previous_node.left == current_node:
                    previous_node.left = current_node.left
                else:
                    previous_node.right = current_node.left
                return root
            else:
                if previous_node is None:
                    return current_node.right
                if previous_node.left == current_node:
                    previous_node.left = current_node.right
                else:
                    previous_node.right = current_node.right
                return root
        if current_node.data < value:
            if current_node.right is not None:
                previous_node = current_node
                current_node = current_node.right
            else:
                print ("The value is not in the tree!")
                return root
        else:
            if current_node.left is not None:
                previous_node = current_node
                current_node = current_node.left
            else:
                print ("The value is not in the tree!")
                return root

bst = delete_node(bst, 'G')
print (bst.min_node())
```
