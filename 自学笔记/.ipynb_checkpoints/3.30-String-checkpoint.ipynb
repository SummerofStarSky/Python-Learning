```python
# 字符串和列表有点像，操作基本相同；
# 注意：对字符串使用索引，只能读不能改
str1 = 'dhjasguidhds'
a = len(str1) # 取字符串长度

b = str1.find('a') # 查找某一个字符，找到返回第一次出现index，没找到则返回-1；同样可以查找某一个短的字符串

# 字符串的分片与列表完全一样
```

```python
# 加法 & 乘法
str1 = 'dhjasguidhds'
str2 = 'nvjdk'

print (str1 + str2) # 字符串 + 字符串， 连接
print (str1 * 3) # 字符串 * 整数， 重复
```

```python
# 分割 和 合并

s1 = 'Feedback is essential in a control system'
s2 = ['Feedback', 'is', 'essential']

a = s1.split(' ')
# 分割结果为列表
b = s1.split('s')
c = s1.split('x')

print (a)
print (b)
print (c)

d = ' '.join(s2)
e = '*'.join(s2)
# 列表合并结果为字符串

print (d)
print (e)
```

```python
# 大小写转换
s1 = 'feedback is essential in a control system'

x = s1.upper() # 全大写
y = x.lower() # 全小写
z = y.capitalize() # 首字母大写

print (x)
print (y)
print (z)
```

```python
# 字符串格式化
# format函数，格式化字符串，也可以格式化数字

s1 = "{}{}{}".format("Feedback", "is", "essential") #按默认顺序
print (s1)
print ("{0} {1} {2}".format('Feedback', 'is', 'essential')) # 设置指定位置
print ('{0} {1} {1} {2}'.format('Feedback', 'is', 'essential')) # 设置指定位置

print ("{:.2f}".format(3.1415926)) # 保留两位小数
print ("{:+.2f}".format(-3.1415926)) # 保留两位小数和符号
print ("{:x}".format(13)) # 转16进制， 用大写输出结果也为大写
print ("{:X}".format(13))
print ("{:o}".format(13)) # 转8进制
```

```python
# 三引号
s1 = '''feedback
is
critical '''

s2 = 'feedback\
is\
critical'

print (s1)
print (s2)
```
