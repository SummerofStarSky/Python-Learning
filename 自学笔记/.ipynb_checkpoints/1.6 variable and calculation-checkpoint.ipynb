```python
a = 100 
b = 12.345
c = 'hello world'
d = True 
letter1 = 'e'
letter2 = 'E'
print (type(a)) # type函数用来输出变量的变量类型
print (type(b))
print (type(c))
print (type(d))
print (chr(a)) # 以下是一些转换变量类型的函数
print (ord(letter1))
print (ord(letter2))
print (str(b))
print (bool(c))
print (int(d))
```

```python
str1 = input()
truenumber = int(str1)
truenumber += 10 #相当于 truenumber=truenumber+10
print (truenumber)
truenumber *= truenumber // 100 #相当于 truenumber=truenumber*（truenumber//100）, //代表整除， %代表取模， **代表指数、求幂
print (truenumber)
a1 = 10
a2 = a1 % 4 # 取模
print (a2)
print (a2 ** a1) # 输出a2的a1次方
```

```python
b1 = 1 == 1 # 比较运算符和逻辑运算符可以直接写在赋值语句中，会将运算结果用于赋值
b2 = 2 > 3
b3 = 45 < 67
b4 = b1 and b2 
b5 = b1 or b3
b6 = not ( b1 and b2 )
print (b1)
print (b2)
print (b3)
print (b4)
print (b5)
print (b6)
print (int(b1))
```

```python
# 求圆的周长和面积
radius = float(input('请输入圆的半径:'))
perimeter = 3.14159 * radius * 2
area = 3.14159 * radius ** 2 # 运算顺序和数学里的一样
print (perimeter)
print (area)
```

```python
# 判断任意给定年份是否是闰年
year = int(input())
testresult = ( year % 4 == 0 and year % 100 != 0 ) or ( year % 400 == 0 )
print (testresult)
```

```python
# 交换两个变量
num1 = 2008
num2 = 6
num3 = num1 
num1 = num2 
num2 = num3
print (num1)
print (num2)
```

```python
# 翻转一个4位数
num = int (input())
p1 = num % 10
num //= 10
p2 = num % 10
num //= 10
p3 = num % 10
p4 = num // 10
newnum = p1 * 1000 + p2 * 100 + p3 * 10 + p4
print (newnum)
```

```python
# 基础必做作业

# 两个数的加法和乘法
a1 = int(input())
a2 = int(input())
print (a1 + a2)
print (a1 * a2)

# 整除和余数
a3 = int(input())
a4 = int(input())
print (a3 // a4)
print (a3 % a4)

# 幂运算
a5b = float(input())
a5i = float(input())
print (a5b ** a5i)
```

```python
# 进阶练习
# 表达式练习
b1 = float(input())
b2 = b1 ** 2 + b1 * 2 + 1 # （b1+1）**2
print (b2)
# 摄氏转华氏温度
cel = float(input())
F = cel * 9 / 5 + 32
print (F)
# 时间转换器
time = int(input())
minutes = time // 60
seconds = time % 60
print (minutes,"分",seconds,"秒")
# 几何小题， 求正方形面积与边长
side = float(input())
area = side ** 2
perimeter = side * 4
print (perimeter)
print (area)
```
